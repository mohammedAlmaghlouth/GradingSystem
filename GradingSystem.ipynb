```python

def DisplayGrades(ML, NumberOfTests):
    GradesFile = open("Grades.txt","r")  # Open the Grades Student file
    ML = []  # Create an Empty Masterist
    for Line in GradesFile:   # Read Each Line in the file
        TL= []                # Create an Empty Student List
        ID,Name, GG = Line.split("#")
        TL.append(ID)
        TL.append(Name)
        for i in GG.split() : # Read Each Grade
            TL.append(i)
        ML.append(TL)         # Add the complete record of the student to the Master List
        if len(TL) > NumberOfTests : # Set the Numer of tests to the maximum
            NumberOfTests = len(TL)
    GradesFile.close()        # Close the File 
        
    # Print the Heading Names and Number of Tests if there is
    print("StudentID   Student Name        ", end =" ")
    for i in range(2,NumberOfTests): 
        print("% 8s" % "Test"+str(i-1), end =" ") 
    print("\n")
    
    # Print Students detailed Inforatmon
    for Line in ML:
        for i in range(len(Line)): 
            if i == 0 :
                print('{:>9}'.format(Line[i]),end = " ") # Print the ID
            elif i == 1: 
                print('{:26}'.format(Line[i]),end = " ") # Print the Name
            else :
                print('{:9}'.format(Line[i]),end = " ") # Print the Test Mark
        print()
    X = input("Press Enter Key to continue...\n") # Wait for any Key to return

    return ML, NumberOfTests
#
# This FUNCTION will do the followings:
# - Receive the MasterList from the main program
# - Print the the details for a student
# 
def DisplayGradeForStudent(ML):
    found = False
    TL = []     # Initialize the a temporary List
    tempID = input("Enter studentID: ")
    for TL in ML :   
        if TL[0] == tempID and not found:  # Check if the Student ID exist
            print("\nStudentID   Student Name        ", end =" ") # Print the headings ID & Name
            for i in range(2,len(TL)):     # Print the headings Tests
                print("% 8s" % "Test"+str(i-1), end =" ") 
            print("\n")
            
            # Print the Student Information
            for i in range(len(TL)): 
                if i == 0 :
                    print('{:>9}'.format(TL[i]),end = " ")
                elif i == 1: 
                    print('{:26}'.format(TL[i]),end = " ")
                else :
                    print('{:9}'.format(TL[i]),end = " ")
            print("\n")
            found = True
            break
    if not found :   # If Student ID does not exist 
        print("\nERROR: Invalid student ID..\n")
        
    X = input("\nPress Enter key to continue...\n")
    return
#
# This Function will do the followings:
# - Traverse throuh the MasterList
# - For each student calculate the Average
# - Return a new List with all Students Average
# 
def AverageTest(ML, NumberOfTests):
    TL = [] # Create a temporary List
    AverageList = [] # Create the Average List
    IN = 0
    if NumberOfTests > 2 :
        print("\nStudentID   Student Name                 Average\n") 
        for TL in ML :  # Traverse therough the Master List
            Total = 0.0
            for i in range(2,len(TL)):   # Average all Test Scores        
                Total = float(TL[i]) + Total
            AV = Total / (NumberOfTests-2)
            AV = round(AV,1)
            AverageList.append(AV)
            print('{:>9} {:26} {:9}'.format(TL[0], TL[1], AverageList[IN])) # Print the Average scores
            IN = IN + 1
    else : 
        print("Students have not taken any test yet")
    X = input("\nPress Enter key to continue...\n")
    return AverageList  # Return a List contains all student Averages
#
# This function will do the followngs:
# - Receive the Master List and create a temporay List for processing
# - Read Student ID and the Quiz number to modify
# - Read the new grade
# - Replace the Old grade with new grade and update the MasterList 
# 
#
def ModifyGrade(ML):
    TL = []
    found = False
    tempID = input("Please Enter studentID: ")
    for TL in ML :
        if TL[0] == tempID :  # If ID Exist in the List
            TestNumber = int(input("Please enter quiz number to modify: "))
            NewGrade = input("Please enter new quiz "+str(TestNumber)+" grade: ")
            found = True 
            if TestNumber <= (len(TL)-2) and TestNumber > 0 :  # if the Quiz Number Valid
                if ValidScore(NewGrade) : # Verify the new entered grade is within the range
                    print()
                    print("Before grade modification: "+TL[0]+"   "+TL[1],end=" ") # Print ID & Name
                    for i in range(2,len(TL)) :        # Print Grade before Modification
                        print("  "+TL[i],end=" ")
                    print()
                    TL.pop(TestNumber+1)
                    TL.insert(TestNumber+1,NewGrade)
                    print("After grade modification:  "+TL[0]+"   "+TL[1],end=" ") # Print ID & Name
                    for i in range(2,len(TL)) :
                        print("  "+TL[i],end=" ")      # Print Grade after modification
                    print()
                else:    
                    print("\nThe new test grade is out of range\n")   
            else:
                print("\nError: Invalid quiz  number")               
    if not found :
        print("\nThe student ID does not exist\n")

    X = input("\nPress Enter key to continue...\n")  
 
    return ML
# This function will do the followings:
# - Receive the masterList with the number of Tests 
# - Traverse through the MasterList
# - Add a new Test for all Students and update the masterList
#
def AddNewTest(ML,NumberOfTests):
    if ML != [] :
        print("Please enter test grades for Test# "+str(NumberOfTests-1)) # Read the next Test
        TL = []         # Create an eampty List
        for TL in ML :  # Traverse through the masterList to go over all students
            NewGrade = input("Please enter grade for student : "+TL[0]+"\n") # Read the new Test Grade for each student
            while not ValidScore(NewGrade) :  # Make sure the Grade is Valid
                NewGrade = input("Please enter a valid grade for student : "+TL[0]+"\n")
            TL.append(str(NewGrade))                # Add the new grade for the student
    else :
        print("\n No Students records found ...")

    X = input("\nPress Enter key to continue...\n")
    
    NumberOfTests = NumberOfTests +1             # Update Number of Tests
    return ML, NumberOfTests
#
# This function will do the followings: 
# - Read the current MasterList and update it with the new Student
# - Request the user to add a new Student ID and 
# - Make sure the ID does not exist in the list and the new student ID should start with 91 and consist of 5 digits
# 
def AddNewStudent(ML,NumberOfTests):
    TL = []
    found = False
    tempID = input("Enter studentID: ")
    if ValidID(tempID) :        # Call a function to verify that the ID is valid
        found, pos = ID_Exist(tempID,ML)
        if found : # Call fuction to make sure that the ID does not exist
            print("ERROR: Student ID is already exist")
        else : 
            TL = []
            TL.append(tempID)
            N=input("Enter the Student Name : ")
            TL.append(N)
            for i in range(2,NumberOfTests) :        
                TIN = input("Enter test"+str(i-1)+" Grade : ")
                while not ValidScore(TIN) :    # Make sure the Grade is valid
                    TIN = input("Enter a Valid test"+str(i-1)+" Grade : ")
                TL.append(str(TIN))
            ML.append(TL)    # append to the MasterList
    else :
        print("Invalid ID, The ID should be 5 digits and sstarts with 91 \n")
        
    X = input("\nPress Enter key to continue...\n")

    return ML
#
# This function will do the followings:
# - Receive the Master List to update
# - Read the student ID that needs to be removed 
# - If Student ID exist, then delete it and update the MasterList
# - If Student ID does not exist, then print a message as below
# 
def DeleteStudent(ML):
    TL = []
    found = False
    tempID = input("Enter studentID: ")
    found, pos = ID_Exist(tempID,ML)  # Call function to verify that the ID exist 
    if found :
        ML.pop(pos)
    else :
        print("ERROR: The Student ID does not exist..")
    X = input("\nPress Enter key to continue...\n")

    return ML
#
# This fuction will do the followings:
# - Write the final output to the file 
# - Return control to the main menue to Exit
#
#
def ExitFunction(ML,NumberOfTests):
    outputFile = open("Grades.txt","w")   # Open the output file
    if ML != [] :
        TL = ML[0]
    else : 
        TL = ML 
    # Print Students Inforatmon
    for Line in ML:
        outputFile.write("% 5s" % Line[0]+"#  "+Line[1].lstrip(" ")+"#  ")
        for i in range(2,len(Line)): 
            outputFile.write('{:9}'.format(Line[i]))
        outputFile.write("\n")
    outputFile.close()
    return
# This fuction will do the following:
# - It is called by more than a function to check if the ID exist or not
# - if the ID exist, then it returns True and the position of the ID
#
def ID_Exist(ID,ML):
    found = False
    TR = []
    pos = 0
    for TR in ML : 
        if ID == TR[0] :
            found = True
            pos = ML.index(TR)   # Read the position of the student record on the MasterList
    return found, pos            # Retrun the results and the position of the record
#
# This fuction will do the followings:
# - Validate if the ID has the format "91XXX", starts with 91 and only 4 digits
#
def ValidID(ID):
    VALD = False
    if len(ID) == 5 and ID[0] == "9" and ID[1] == "1" and ID.isdigit():
        VALD = True
    return VALD 
#
# This fuction will do the followings:
# - Validate if the score is digit and falls between 0 and 100
#
def ValidScore(Score):
    VALD = False
    try : 
        X = float(Score)
        if float(Score)>=0 and float(Score)<=100 :
            VALD = True
    except ValueError: 
        VALD = False
        
    return VALD 

#
# Main Function will do the follwings:
# - Initialize the Master List and Number of Tests. 
# - Display the main menue as shown below
# - Make sure that the user enter a valid option
# - If the user enters option 8, the program will write the output file and exit the program.
#
def main():
    MasterList = []         # Master List contains stuedents IDs, Names, Tests...
    NumberOfTests = 0       # Number of Tests is set to 0 initially
    OP = "1"
    while OP != "8" : 
        if OP == "1" or OP == "2" or OP == "3" or OP == "4" or OP == "5" or OP == "6" or OP == "7" :
            print("1. Display Grade Info for all students")
            print("2. Display Grade Info for a particular student")
            print("3. Display tests average for all students")
            print("4. Modify a particular test grade for a particular student")
            print("5. Add test grades for a particular test for all studentss")
            print("6. Add a new Student")
            print("7. Delete a student")
            print("8. Exit")
            print()
            OP = input("Please select your choice: ")
            print()
            if OP == "1" :
                MasterList, NumberOfTests = DisplayGrades(MasterList, NumberOfTests)
            elif OP == "2" :
                DisplayGradeForStudent(MasterList)
            elif OP == "3" :
                AverageList = AverageTest(MasterList,NumberOfTests)
            elif OP == "4":
                MasterListr = ModifyGrade(MasterList) 
            elif OP == "5" :
                MasterListr,NumberOfTests = AddNewTest(MasterList,NumberOfTests)
            elif OP == "6":
                MasterList = AddNewStudent(MasterList,NumberOfTests) 
            elif OP == "7" :
                MasterList = DeleteStudent(MasterList)                
        else :
            print()
            print("An invalid choice is entered")
            print()
            OP = input("Please select your choice: ")
            print()
    ExitFunction(MasterList,NumberOfTests)  # call ExistFunction to write the ouptfile and exit. 
     
#
# Call the Main Function to star the program
#
main()
    
```
